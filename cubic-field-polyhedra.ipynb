```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

%matplotlib widget
```

```python
def add_polyhedron(axes,verts,colour,mark='o',hatch='.'):
    hull = ConvexHull(verts)
    if hatch != None:
        hatch = 5*hatch
    for s in hull.simplices:
        tri = Poly3DCollection([verts[s]],hatch=hatch)
        tri.set_color(colour)
        tri.set_alpha(0.5)
        axes.add_collection3d(tri)
    axes.scatter(verts[:,0],verts[:,1],verts[:,2],marker=mark,color='black',hatch=hatch)
```
